```python
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm 
# that groups together points that are closely packed together, while marking points that lie alone 
# in low-density regions as outliers. It requires two parameters: epsilon (maximum distance between 
# two samples to be considered neighbors) and min_samples (minimum number of points to form a dense region).

# Import necessary libraries
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate a 2D dataset of 1000 points shaped like two interleaving half circles (moons) with added noise
X, y = make_moons(n_samples=1000, noise=0.05)

# Create a DBSCAN instance with an epsilon of 0.05 and a minimum of 5 samples to form a dense region
dbscan = DBSCAN(eps=0.05, min_samples=5)

# Fit the DBSCAN model to the dataset
dbscan.fit(X)

# Access the cluster labels assigned by the algorithm (-1 indicates noise points)
labels = dbscan.labels_
print("These are all the labels for each instances. A -1 value implies anomalies for the algorithm.")
print(labels)
```

```python
print(f"The indices of the core instances are {len(dbscan.core_sample_indices_)}")
print(dbscan.core_sample_indices_)
```

```python
print("The core instances themselves are available in the components_ instances variable")
print(dbscan.components_)
```

```python
# Plot the clusters
plt.figure(figsize=(8, 6))
# Use a scatter plot with a colormap. Points labeled -1 (noise) will be plotted in black.
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="Paired", s=10)
plt.title("DBSCAN Clustering of make_moons Dataset")
plt.xlabel("X")
plt.ylabel("X")
plt.grid(True)
plt.show()
```

```python
# The result is less than optimal. We can improve the clustering algorithm by increasing eps to 0.2

# Create a DBSCAN instance with an epsilon of 0.2 and a minimum of 5 samples to form a dense region
dbscan2 = DBSCAN(eps=0.2, min_samples=5)

# Fit the DBSCAN model to the dataset
dbscan2.fit(X)

# Access the cluster labels assigned by the algorithm (-1 indicates noise points)
labels_dbscan2 = dbscan2.labels_

# Plot the clusters
plt.figure(figsize=(8, 6))
# Use a scatter plot with a colormap. Points labeled -1 (noise) will be plotted in black.
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan2, cmap="Paired", s=10)
plt.title("DBSCAN Clustering of make_moons Dataset with epsilon equals 0.2")
plt.xlabel("X")
plt.ylabel("X")
plt.grid(True)
plt.show()

# We obtain a much better result.
```

```python
# The DBSCAN class does not have a predict method, so it cannot predict
# if a new instance belongs to a particular cluster
# We are going to implement this feature by training a KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier
import numpy as np

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan2.components_, dbscan2.labels_[dbscan2.core_sample_indices_])

# Let's try to predict to which cluster this new instances belong to using the model trained
X_new = np.array([[-0.5,0], [0,0.5], [1,-0.1], [2,1]])
print(f"The predicted clusters for the new instances are: {knn.predict(X_new)}")
print(f"The probability of this predictions are:\n {knn.predict_proba(X_new)}")
```

```python
# Since there is no anomaly in the training set, the classifier always chooses a cluster
# We are going to introduce a maximum distance, in which case the two instances that are
# far away from both clusters are classified as anomalies

# Find the nearest core point (distance and index) for each new instance in X_new
# knn.kneighbors returns the distances and indices of the nearest neighbors
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)

# Use the indices to get the predicted cluster labels from DBSCAN's core samples
# Only core sample indices are considered to simulate DBSCAN's density-based classification
y_pred = dbscan2.labels_[dbscan2.core_sample_indices_][y_pred_idx]

# Mark any points whose nearest core point is farther than eps (0.2) as noise (label = -1)
y_pred[y_dist > 0.2] = -1

# Print the simulated DBSCAN prediction for new instances
print(f"The pew prediction is: {y_pred.ravel()}")
```

```python
# Plot the original dataset clustered by DBSCAN
plt.figure(figsize=(10, 7))

# Scatter plot of original data colored by DBSCAN cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan2, cmap="Paired", s=10, label="Original data")

# Plot new instances with red crosses
plt.scatter(X_new[:, 0], X_new[:, 1], marker='x', c='red', s=100, linewidth=2, label="New instances")

# Aesthetics
plt.title("DBSCAN Clustering of make_moons Dataset with New Instances")
plt.xlabel("X")
plt.ylabel("X")
plt.grid(True)
plt.legend()
plt.show()
```
