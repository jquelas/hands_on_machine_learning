```python
# Dimensionality Reduction Visualization: Swiss Roll Dataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll

# Set random seed for reproducibility
np.random.seed(42)

# Generate the Swiss Roll dataset: 3D spiral-shaped surface with noise
X, t = make_swiss_roll(n_samples=1000, noise=0.2)

# Create a 3D plot to visualize the Swiss Roll manifold
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot: color each point by its unrolled position (t parameter)
ax.scatter(
    X[:, 0], X[:, 1], X[:, 2],
    c=t,                           # Coloring based on position along the roll
    cmap=plt.get_cmap("jet"),      # Use the jet colormap for vivid gradients
    edgecolor='k',                 # Add black edges to each point
    alpha=0.75                     # Slight transparency for better depth perception
)

# Set axis labels for the 3D plot
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Add a title to the plot
ax.set_title("Swiss Roll Dataset")

# Adjust the view angle: elev=5 gives a low elevation, azim=100 adds X-axis tilt (~20°)
ax.view_init(elev=5, azim=100)

# Display the final plot
plt.show()
```

```python
from sklearn.manifold import Isomap

# Isomap: Nonlinear dimensionality reduction technique
# Attempts to "unroll" the Swiss Roll by preserving geodesic distances
isomap = Isomap(n_components=2, n_neighbors=10)  # 2D embedding, using 10 neighbors for graph construction
X_isomap = isomap.fit_transform(X)  # Transform the 3D data into a 2D representation

# Plot the unrolled representation of the Swiss Roll
plt.figure(figsize=(8, 6))
plt.scatter(
    X_isomap[:, 0], X_isomap[:, 1],
    c=t,                    # Color points by their original position on the roll (t)
    cmap="jet",             # Use jet colormap for smooth gradient
    edgecolor="k",          # Add black edges to points
    alpha=0.75              # Slight transparency for visual clarity
)

# Axis labels and title
plt.xlabel("Unrolled Dimension 1")
plt.ylabel("Unrolled Dimension 2")
plt.title("Unrolled Swiss Roll using Isomap")

# Add colorbar to map colors back to original t-values
plt.colorbar(label="t values")

# Display the final plot
plt.show()
```
